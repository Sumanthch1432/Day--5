```python
import seaborn as sns
```

```python
# Set seaborn style
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
print("\n--- Dataset Info ---")
print(df.info())
```

```python
print("\n--- Statistical Summary ---")
print(df.describe())
```

```python
print("\n--- Statistical Summary ---")
print(df.describe())
```

```python
print("\n--- Value Counts for 'Survived' ---")
print(df['Survived'].value_counts())
```

```python
# Visualization 1: Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Survival Count
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 3: Boxplot of Age by Survived
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df, palette='Set3')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# Visualization 4: Correlation Heatmap
plt.figure(figsize=(10,8))
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Visualization 5: Pairplot (Cleaned Data)
# Clean data first
cleaned_df = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].dropna()
cleaned_df = cleaned_df.astype(float)
```

```python
# Pairplot
pairplot = sns.pairplot(cleaned_df, hue='Survived', palette='husl')
pairplot.fig.suptitle('Pairplot of Titanic Data (Cleaned)', y=1.02)
plt.show()
```
